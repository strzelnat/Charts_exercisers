```python
import pandas as pd
import numpy as np
import chart_studio.tools as tl
import chart_studio.plotly as pl
import plotly.graph_objects as go
```

```python
df = pd.read_csv('athlete_events.csv')
df['Medal'] = df['Medal'].fillna('none')
```

```python
df['Gold'] = df['Medal'].apply(lambda x: 1 if x == 'Gold' else 0)
df['Silver'] = df['Medal'].apply(lambda x: 1 if x == 'Silver' else 0)
df['Bronze'] = df['Medal'].apply(lambda x: 1 if x == 'Bronze' else 0)
df['none'] = df['Medal'].apply(lambda x: 1 if x == 'none' else 0)
```

```python
df.head()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

medal_colors = {
    'Gold': 'red',
    'Silver': 'green',
    'Bronze': 'blue'
}

fig = make_subplots(
    rows=1,
    cols=2,
    subplot_titles=("Female", "Male"),
    shared_yaxes=True
)

sex_to_col = {'F': 1, 'M': 2}

for sex in ['F', 'M']:
    for medal in ['Gold', 'Silver', 'Bronze']:
        filtered_data = df[(df['Sex'] == sex) & (df[medal] == 1)]
        fig.add_trace(
            go.Box(
                y=filtered_data['Age'],
                name=medal,
                boxmean=True,
                boxpoints='outliers',
                marker_color = medal_colors[medal]
            ),
            row=1,
            col=sex_to_col[sex]
        )

fig.update_layout(
    title='Age Distribution by Medal Type and Sex',
    yaxis_title='Age',
    template='plotly_dark',
    showlegend=True
)

fig.show()
```

```python
pl.plot(fig, filename= 'box_plot_medals', auto_open = True)
```

```python
import plotly.express as px
```

```python
fig = px.scatter(
    df.query("Medal == 'Gold' & Year == 2012"),
    x='Height',
    y='Weight',
    color='Sex',
    hover_data = ['Sex', 'Year', 'Games'],
    facet_col = 'Sex',
    facet_col_spacing= 0.2,
    trendline = 'ols',
     trendline_color_override="white"

)




fig.update_layout(template = 'plotly_dark', title = 'Height vs Weight of Gold Medalists at the 2012 Olympics by Sex (with Trendline)')

fig.show()
```

```python
pl.plot(fig, filename= 'Height vs Weight', auto_open = True)
```

```python
fig = px.parallel_categories(df.query('Year == 2012 & Age <= 26 '),
                             dimensions = ['Sex', 'Age', 'Medal'],
                             template = 'ggplot2',
                             color = 'Age'
                             )

fig.update_layout(
    title = 'Medal Distribution by Age and Sex – London 2012 Olympics'
)

fig.show()
```
